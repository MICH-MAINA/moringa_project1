**4. Data Cleaning**

```python
#Dropping columns with many missing values

aviation_data.drop(columns=['Latitude','Longitude', 'Schedule', 'Air.carrier', 'Airport.Code','Airport.Name' ], inplace=True)
```

```python
aviation_data.info()
```

```python
#Add new column
aviation_data['total_injuries'] = aviation_data['Total.Fatal.Injuries'] + aviation_data['Total.Serious.Injuries'] + aviation_data['Total.Minor.Injuries']

#Create a new column for year to get year trend
aviation_data['Year'] = pd.DatetimeIndex(aviation_data['Event.Date']).year

aviation_data.head()
```

```python
#replace null with 0 total.Fatal.Injuries       Total.Serious.Injuries  Total.Minor.Injuries    Total.Uninjured

aviation_data[['Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries', 'Total.Uninjured', 'Number.of.Engines', 'total_injuries']] = aviation_data[['Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries', 'Total.Uninjured', 'Number.of.Engines', 'total_injuries']].fillna(0)

aviation_data.head()
```

```python
#Delete rows where event ID is null
aviation_data = aviation_data.dropna(subset=['Event.Id', 'Model', 'Aircraft.Category'])
aviation_data.info()
```

```python
# Convert date columns
aviation_data['Event.Date'] = pd.to_datetime(aviation_data['Event.Date'], errors='coerce')
aviation_data['Publication.Date'] = pd.to_datetime(aviation_data['Publication.Date'], errors='coerce')
aviation_data.head()
```

```python
risk_summary = aviation_data.groupby('Make').agg({
    'Event.Id': 'count',
    'Total.Fatal.Injuries': 'sum',
    'total_injuries': 'sum'
}).rename(columns={'Event.Id': 'accident_count'}).sort_values(by='accident_count', ascending=False)
risk_summary.head()
```

```python
aviation_data.info()
```

Replacing all the fatal(*) with Fatal

```python
aviation_data['Injury.Severity'] = aviation_data['Injury.Severity'].str.upper()
aviation_data['Injury.Severity'] = aviation_data['Injury.Severity'].replace(
    to_replace=r'FATAL\(\d+\)|FATAL', value='FATAL', regex=True)

aviation_data.head()
```

```python
#Copy of the csv to export
aviation_data.to_csv('aviation_data_cleaned.csv', index=False)

```
